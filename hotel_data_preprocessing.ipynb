```python
import nltk,re
from nltk.stem import WordNetLemmatizer, PorterStemmer
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
import pandas as pd

data = pd.read_csv(r'C:\Python_Labs\AI Bootcamp 2025\tripadvisor_hotel_reviews (1).csv')
```

```python
data.groupby('Rating').apply(lambda x: x.Review.iloc[0])
```

```python
data.head()
```

```python
#create a new olumn for lowercase data 
data['lower'] = data['Review'].str.lower()
data.head()
```

```python
#new column for no stop words
stop = stopwords.words('english')
stop.remove('not')
data['no_stop_words'] = data['lower'].apply(lambda x: ' '.join([word for word in x.split() if word not in (stop)]))
data.head()
```

```python
#convert * to word star in a no punct column using regular expression re.sub() function
data['no_punct'] = data['no_stop_words'].apply(lambda x: re.sub(r'[*]',' star',x))
data.head()
```

```python
#update the no_punct column to remove all non-alphabetic characters
data['no_punct'] = data.apply(lambda x: re.sub(r"([^\w\s])", ' ', x['no_punct']), axis=1)
data['no_punct'].head()
```

```python
data['tokens'] = data['no_punct'].apply(word_tokenize)
data['tokens'].head()
```

```python
stemmer = PorterStemmer()
data['stemmed'] = data['tokens'].apply(lambda x: [stemmer.stem(word) for word in x])
data.head()
```

```python
lemmatizer = WordNetLemmatizer()
data['lemmatized'] = data['tokens'].apply(lambda x: [lemmatizer.lemmatize(word) for word in x])
data.head()
```

```python
token_clean = sum(data['lemmatized'],[])

unigrams = (pd.Series(nltk.ngrams(token_clean,1)).value_counts())
unigrams
```

```python
bigrams = (pd.Series(nltk.ngrams(token_clean, 2)).value_counts())
bigrams
```
